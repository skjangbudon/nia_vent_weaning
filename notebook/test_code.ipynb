```python
# check final 

import joblib
import pandas as pd, numpy as np
from sklearn.metrics import confusion_matrix, roc_auc_score

model_path = '/VOLUME/nia_vent_weaning/model/RF.pkl'
data_path = '/VOLUME/nia_vent_weaning/output/input/testset.csv'

# import test dataset
testset = pd.read_csv(data_path)
test_x = testset.drop(['pid', 'label'], axis=1)
test_y = list(testset['label'])

# import model
clf = joblib.load(model_path) 

# calculate AUC score
y_prob = clf.predict_proba(test_x)[:,1]
pred_positive_label = y_prob
y_pred = np.where(y_prob > 0.5, 1, 0)

roc_score = roc_auc_score(test_y, pred_positive_label, average='micro')
print('FINAL PREDICTION AUC : ', round(roc_score,2))
```

```python
testset['y_prob'] = y_prob
testset['y_pred'] = y_pred
testset['Real'] = test_y
result_df = testset[['pid', 'y_prob', 'y_pred', 'Real']]
result_df.to_csv('./pred_output.csv')
result_df
```
